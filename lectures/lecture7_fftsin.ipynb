```python
import os
import numpy as np
import pylab as plt
import scipy.signal as sg
import sounddevice as sd
import pandas as pd
from time import sleep
from sklearn.model_selection import train_test_split

plt.rcParams['axes.grid'] = True
```

```python
fig, ax = plt.subplots(2, figsize=(4,4), dpi=150)
plt.tight_layout()

N = 30
l = 1
n = np.arange(N)
x = np.sin(2*np.pi*n*l/N)
X = np.fft.fft(x)



ax[0].stem(x, basefmt='k')
ax[0].set_xlabel('k')
ax[1].stem(X.imag, linefmt='r', markerfmt='or', basefmt='None')
ax[1].stem(X.real, linefmt='b', markerfmt='xb', basefmt='None')
ax[1].xlabel('k')
```
